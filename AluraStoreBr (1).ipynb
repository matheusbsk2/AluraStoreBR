### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
import matplotlib.pyplot as plt

faturamento_loja1 = sum(loja['Preço'])
faturamento_loja2 = sum(loja2['Preço'])
faturamento_loja3 = sum(loja3['Preço'])
faturamento_loja4 = sum(loja4['Preço'])

faturamento = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
```

```python
plt.figure(figsize=(8, 5))
plt.bar(lojas, faturamento, color=['blue', 'green', 'orange', 'red'])
plt.title('Faturamento das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.show()
```

# 2. Vendas por Categoria


```python
import matplotlib.pyplot as plt

categorias_loja1 = loja['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
plt.barh(categorias_loja1.index, categorias_loja1.values, color='blue')
plt.title('Categorias Mais Vendidas - Loja 1')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja2.index, categorias_loja2.values, color='green')
plt.title('Categorias Mais Vendidas - Loja 2')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja3.index, categorias_loja3.values, color='orange')
plt.title('Categorias Mais Vendidas - Loja 3')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(categorias_loja4.index, categorias_loja4.values, color='red')
plt.title('Categorias Mais Vendidas - Loja 4')
plt.xlabel('Quantidade de Produtos Vendidos')
plt.ylabel('Categoria')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_loja1 = loja[['Produto', 'Quantidade de parcelas']]
produtos_loja2 = loja2[['Produto', 'Quantidade de parcelas']]
produtos_loja3 = loja3[['Produto', 'Quantidade de parcelas']]
produtos_loja4 = loja4[['Produto', 'Quantidade de parcelas']]

produtos_loja1_sum = {}
for produto, quantidade in zip(produtos_loja1['Produto'], produtos_loja1['Quantidade de parcelas']):
    if produto in produtos_loja1_sum:
        produtos_loja1_sum[produto] += quantidade
    else:
        produtos_loja1_sum[produto] = quantidade

produtos_loja2_sum = {}
for produto, quantidade in zip(produtos_loja2['Produto'], produtos_loja2['Quantidade de parcelas']):
    if produto in produtos_loja2_sum:
        produtos_loja2_sum[produto] += quantidade
    else:
        produtos_loja2_sum[produto] = quantidade

produtos_loja3_sum = {}
for produto, quantidade in zip(produtos_loja3['Produto'], produtos_loja3['Quantidade de parcelas']):
    if produto in produtos_loja3_sum:
        produtos_loja3_sum[produto] += quantidade
    else:
        produtos_loja3_sum[produto] = quantidade

produtos_loja4_sum = {}
for produto, quantidade in zip(produtos_loja4['Produto'], produtos_loja4['Quantidade de parcelas']):
    if produto in produtos_loja4_sum:
        produtos_loja4_sum[produto] += quantidade
    else:
        produtos_loja4_sum[produto] = quantidade
```

```python
print("Produtos mais vendidos Loja 1:")
print(sorted(produtos_loja1_sum.items(), key=lambda x: x[1], reverse=True)[:5])
print("\nProdutos menos vendidos Loja 1:")
print(sorted(produtos_loja1_sum.items(), key=lambda x: x[1])[:5])

print("\nProdutos mais vendidos Loja 2:")
print(sorted(produtos_loja2_sum.items(), key=lambda x: x[1], reverse=True)[:5])
print("\nProdutos menos vendidos Loja 2:")
print(sorted(produtos_loja2_sum.items(), key=lambda x: x[1])[:5])

print("\nProdutos mais vendidos Loja 3:")
print(sorted(produtos_loja3_sum.items(), key=lambda x: x[1], reverse=True)[:5])
print("\nProdutos menos vendidos Loja 3:")
print(sorted(produtos_loja3_sum.items(), key=lambda x: x[1])[:5])

print("\nProdutos mais vendidos Loja 4:")
print(sorted(produtos_loja4_sum.items(), key=lambda x: x[1], reverse=True)[:5])
print("\nProdutos menos vendidos Loja 4:")
print(sorted(produtos_loja4_sum.items(), key=lambda x: x[1])[:5])
```

# 5. Frete Médio por Loja


```python
frete_loja1 = loja['Frete']
frete_loja2 = loja2['Frete']
frete_loja3 = loja3['Frete']
frete_loja4 = loja4['Frete']

soma_frete_loja1 = 0
for frete in frete_loja1:
    soma_frete_loja1 += frete
media_frete_loja1 = soma_frete_loja1 / len(frete_loja1)

soma_frete_loja2 = 0
for frete in frete_loja2:
    soma_frete_loja2 += frete
media_frete_loja2 = soma_frete_loja2 / len(frete_loja2)

soma_frete_loja3 = 0
for frete in frete_loja3:
    soma_frete_loja3 += frete
media_frete_loja3 = soma_frete_loja3 / len(frete_loja3)

soma_frete_loja4 = 0
for frete in frete_loja4:
    soma_frete_loja4 += frete
media_frete_loja4 = soma_frete_loja4 / len(frete_loja4)
```

```python
print("Custo médio de frete Loja 1: R$", round(media_frete_loja1, 2))
print("Custo médio de frete Loja 2: R$", round(media_frete_loja2, 2))
print("Custo médio de frete Loja 3: R$", round(media_frete_loja3, 2))
print("Custo médio de frete Loja 4: R$", round(media_frete_loja4, 2))
```

# Introdução
Este relatório tem como objetivo apresentar uma análise detalhada do desempenho das quatro lojas fictícias da Alura Store, com o intuito de fornecer uma recomendação sobre qual loja o Senhor João deve vender. A partir de dados sobre faturamento total, categorias de produtos mais e menos vendidas, avaliações dos clientes, produtos mais e menos vendidos e custos de frete, buscaremos identificar qual loja tem o melhor desempenho global, considerando tanto os aspectos financeiros quanto a satisfação dos clientes.
